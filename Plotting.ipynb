```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
plt.style.use('seaborn-whitegrid')
clrs = sns.color_palette('husl', n_colors=5)  # a list of RGB tuples
```

```python
ticks = [1, 6, 20]
labels = ['read_fast', 'read_columnar', 'read_csv (chunked)', 'read_csv']
read_fast = [8.66, 6.58, 2.66]
read_columnar = [66, 127, 398]
read_cc = [140, 132, 119]
read_csv = [175, 201, 178]
cols = [read_fast, read_columnar, read_cc, read_csv]

plt.figure(figsize=(8,6))
ind = np.arange(3) 
width = 0.15
for i, col in enumerate(cols):
    plt.bar(ind + width * i, col, width, label=labels[i], color=clrs[i])

plt.title("Execution Time vs Number of Column Filters Applied (CA_police dataset)")

plt.xticks(ind + width, ticks)
plt.ylabel('Execution Time (s)')
plt.xlabel('Number of Columns Filtered on')

plt.legend(loc='best')
plt.show()
```
